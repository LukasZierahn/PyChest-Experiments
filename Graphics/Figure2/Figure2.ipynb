```python
import numpy as np
import matplotlib.pyplot as plt


fig, ax = plt.subplots(figsize=(8,5))
ax.set_title(f'Piece-wise Hidden Irrational Rotation')
ax.set(xlabel='Time', ylabel='')

sequence = np.genfromtxt("sequences/Hidden_Rotary_overlap_0.1/0/5000/data.csv")
estimate = np.genfromtxt(f"sequences/Hidden_Rotary_overlap_0.1/0/5000/changepoints.csv")

n = 10
sample = np.arange(len(sequence)//n)
ax.plot(sample * n, sequence[sample * n])

ax.axvline(estimate * len(sequence), color="red", ls="--")


fig.savefig(f"Figure2.png", dpi=1200)
```
