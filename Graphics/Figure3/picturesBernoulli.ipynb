```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
data = pd.read_csv("Bernoulli_1chpt/errors.csv")
data
```

```python
data["Unnamed: 2"][2:].to_numpy()
```

```python
plt.figure(figsize=(12, 8))

plt.ylim(top=.5)

plt.ylabel("estimation error", fontsize=18)
plt.xlabel("sample length", fontsize=18)

plt.plot([0, *data["Unnamed: 0"][2:11]], [1, *data.Bernoulli[2:11].to_numpy(np.float)], "--o", label="PyChEst", color="red", linewidth=3.0, ms=10.0)
plt.plot([0, *data["Unnamed: 0"][2:11]], [1, *data["Unnamed: 2"][2:11].to_numpy(np.float)], "--o", label="wbs", linewidth=3.0, ms=10.0)

plt.title("Piece-wise i.i.d. Bernoulli", fontsize=22)

plt.legend()
plt.savefig("Bernoulli.png", dpi=100)
plt.show()
```
